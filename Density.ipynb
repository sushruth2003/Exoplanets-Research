```python
import pandas as pd
import numpy as np
import array
import math
AllPlanets = pd.read_csv("/Users/sushruth/Desktop/ASDRP/AllTypesOrbsMax2.csv")
pd.set_option('display.max_rows', None)
for i in range (0, AllPlanets.shape[0]):
    orbper = AllPlanets.loc[i,'pl_orbper']
    orbsmax = AllPlanets.loc[i,'pl_orbsmax']
    star = AllPlanets.loc[i,'st_mass']
    upper = 4*((float(math.pi))**2)*(((float(orbsmax))*1.496*(10**11))**3)
    lower = ((float(orbper)*86400)**2)*6.67408*(10**(-11))
    prelim = float(upper)/float(lower)
    mass = float(prelim)-(float(star)*1.989*(10**30))
    mass = mass *1000
    rade = AllPlanets.loc[i,'pl_rade']
    rad = ((float(rade)*6.371*10**8)**3)*(4/3)*float(math.pi)
    density = mass/rad
    print(density)
```
